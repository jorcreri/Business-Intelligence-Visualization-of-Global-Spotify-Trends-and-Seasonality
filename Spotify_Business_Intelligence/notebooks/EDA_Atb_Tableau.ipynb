# Limpieza del dataset "attributes_popularity" para visualización en Tableau.

```python
from pathlib import Path
import pandas as pd, numpy as np
pd.set_option('display.max_columns', 100)
DATA_DIR = Path.cwd().parent / "data" 
SPOTIFY = DATA_DIR / 'attributes_popularity.csv'

if not SPOTIFY.exists():
    raise FileNotFoundError(f"Training file not found at {TRAIN_CSV}. Please place cs-training.csv in data/ and re-run.")

atb = pd.read_csv(SPOTIFY)

print('Datos cargados correctamente.')
```

**Explicación:** En esta celda se cargará el .csv "attributes_popularity" (sacado de Kaggle) para modelizarlo a nuestro gusto y mantener la información que nos resulte relevante.

```python
atb.info()
atb.head(7)
```

**Explicación:** Observamos la información principal y la cabecera del dataset original.

```python
cols_to_keep = [
    'artists', 
    'track_name', 
    'popularity', 
    'duration_ms', 
    'danceability', 
    'energy', 
    'loudness', 
    'speechiness', 
    'acousticness', 
    'instrumentalness', 
    'liveness', 
    'valence', 
    'tempo', 
    'track_genre'
]

atb = atb[cols_to_keep]

print(f"Dimensiones: {atb.shape}")
atb.head()
```

**Explicación:** En esta celda cargamos el archivo y nos deshacemos inmediatamente de las columnas "basura" o demasiado técnicas (id, key, mode, signature) que ocuparían memoria y espacio en Tableau sin aportar valor a las gráficas de emociones o popularidad.

```python
atb['duration_min'] = round(atb['duration_ms'] / 60000, 2)

atb = atb.drop(columns=['duration_ms'])
atb = atb.drop_duplicates()
atb = atb.dropna()

print(f"Dimensiones tras limpieza básica: {atb.shape}")
atb.head()
```

**Explicación:** Aquí convertimos la duración a minutos (nadie visualiza canciones en milisegundos) y aseguramos la calidad del dato eliminando duplicados exactos y valores vacíos que podrían romper los cálculos de promedios en Tableau.

```python
top_genres = atb['track_genre'].value_counts().nlargest(25).index.tolist()
print("Géneros seleccionados para el análisis:", top_genres)

atb_clean = atb[atb['track_genre'].isin(top_genres)].copy()
print(f"Dimensiones finales tras filtrar géneros: {atb_clean.shape}")
```

**Explicación:** Si intentamos meter 114 géneros en un gráfico de cajas o un mapa de árbol, Tableau se vuelve ilegible y lento (además que géneros secundarios son irrelevantes). Al quedarnos con los Top 25 (Samba, Mandopop, Tango, etc.), garantizamos que las gráficas claras.

```python
OUTPUT_FILE = DATA_DIR / 'atb_pop_limpio.csv'

atb_clean.to_csv(OUTPUT_FILE, index=False, encoding='utf-8')

print(f"Archivo guardado en: {OUTPUT_FILE}")
```

**Explicación:** Guardamos el nuevo dataset como .csv para exportar a Tableau y comenzar con la parte visual.
